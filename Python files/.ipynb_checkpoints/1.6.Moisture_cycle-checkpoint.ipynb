## 1. Regional moisture recycling metrics for a single grid cell

```python
link_path = os.getcwd()
# fixing local link_path, change here where you installed the link dataset mentioned below
ENVM1502_link_path = os.link_path.dirname(os.link_path.dirname(os.link_path.dirname(link_path)))
link_link_path = f'{ENVM1502_link_path}\\Unit 1\\Ex6-Moisture recycling\\Link2019'
```

```python
# load tha source-receptor data based on 1.5degree grid cells 
# + some data manipulation to get the data back on the right location on the globe

cell_longitudes  = np.load(link_link_path+r"/Further data/cell_longitudes.npy") 
cell_latitudes   = np.load(link_link_path+r"/Further data/cell_latitudes.npy")
considered_cells = np.load(link_link_path+r"/Further data/considered_cells.npy")

# Defining the real coordinates of the land grid cells (coordinates refer to the center of a cell)
coordinates_land_grid_cells = np.zeros(considered_cells.shape)
for x in range(len(considered_cells[0,:])): 
    arr_coordinate_lat = considered_cells[0][x]   
    latitude_real = cell_latitudes[arr_coordinate_lat]
    coordinates_land_grid_cells[0,x] = latitude_real

for x in range(len(considered_cells[1,:])): 
    arr_coordinate_lon = considered_cells[1][x]
    longitude_real = cell_longitudes[arr_coordinate_lon]
    if longitude_real > 180:
        longitude_real = longitude_real - 360 # longitudes should range from -180 to 180 and not from 0 to 360
    coordinates_land_grid_cells[1,x] = longitude_real 

coordinates_land_grid_cells_list = []
for x in range(len(coordinates_land_grid_cells[0,:])):
    the_coordinate = {'lat': coordinates_land_grid_cells[0,x], 'lon': coordinates_land_grid_cells[1,x]}
    coordinates_land_grid_cells_list.append(the_coordinate)

# now we have created the coordinates of all land grid cells from which evaporation was tracked
print(coordinates_land_grid_cells_list[0:2],coordinates_land_grid_cells_list[-2:])
```

write a function to find nearest gridcell

```python
def distances(data, v):
    p = math.pi/180
    diam_earth = 12742 # km
    the_distances = []
    lat1 = v['lat']
    lon1 = v['lon']
    for a in range(len(data)):
        lat2 = data[a]['lat']
        lon2 = data[a]['lon']
        # Haversine formula
        hav = 0.5 - cos((lat2-lat1)*p)/2 + cos(lat1*p)*cos(lat2*p) * (1-cos((lon2-lon1)*p)) / 2
        the_distances.append(diam_earth * asin(sqrt(hav)))
    return the_distances 
```

We are intersted in the hudson catchment, around -70 lat, 40 lon, we'll take the coordinates of our gauge

```python
# insert our point of interest
POI = {'lat':42.75222, 'lon': -73.68889}

# find the closest index of the land cells
the_distances = distances(coordinates_land_grid_cells_list, POI)
closest_value = min(the_distances)
x_cell = the_distances.index(closest_value)
closest_value_coordinates = coordinates_land_grid_cells_list[x_cell]
print('POI = \t\t' + str(POI))
print('closest = \t' + str(closest_value_coordinates))
print('distance = \t' + str(round(the_distances[x_cell],2)) + ' km')
print('index closest = ' + str(x_cell))
```

```python
# having the closest index, we will retrieve the fate of evaporation from that specific grid cell

# Loading of the source-receptor matrix - (land grid cell to grid)    
Era_Int_2001_2018_cell           = np.load(link_link_path+r"/Matrices/Era_Int_2001_2018_matrix_yr.npy")

#Loading of the total inputs (row 0 = Jan, row 1 = Feb, ..., row 11 = Dec, row 12 = yr)
Era_Int_2001_2018_Input_mm       = np.load(link_link_path+r"/Further Data/Era_Int_2001_2018_Final_Inputs_mm.npy")

# Loading of the relevant losses matrix
Era_Int_2001_2018_Losses_percent = np.load(link_link_path+r"/Further Data/Losses/Era_Int_2001_2018_Losses_Percent_yr_corr.npy")

# Create a reshaped array for the projection on the map 
Considered_cell = Era_Int_2001_2018_cell[:,[x_cell]]
Reshaped_array_cell = Considered_cell.reshape(107,240)

# Defining x,y and z values
longitudes_cells=np.tile(cell_longitudes,(107,1))
latitudes_cells=np.transpose(np.tile(cell_latitudes,(240,1)))     
values_yr_percent_cell=Reshaped_array_cell/np.sum(Reshaped_array_cell)*100

# index of the considered cell 
ind_lat = np.where(cell_latitudes == closest_value_coordinates['lat'])
cell_longitudes_neg = cell_longitudes
cell_longitudes_neg[cell_longitudes > 180 ] = cell_longitudes[cell_longitudes > 180 ] - 360
ind_lon = np.where(cell_longitudes == closest_value_coordinates['lon'])
print(ind_lat)
print(ind_lon)
```

First look globally

```python
# From which we can create a global plot (similar to the viewer, but with different color scale)
# TO DO reminder for the lecturer: fix the colorbar and colorbar ticks 
# (but fine for now as it is not the most important issue)

def global_figure(data, low, high, label):
    # start the figure
    fig = plt.figure(figsize=(11,8.5))

    # Set the axes using the specified map projection
    ax=plt.axes(projection=ccrs.PlateCarree())

    # Add cyclic point to data
    data, lons = add_cyclic_point(data, longitudes_cells[0,:])
    lats = latitudes_cells[:,0]

    # Make a filled contour plot
    cs = ax.contourf(lons, lats, data, 20,
                     transform = ccrs.PlateCarree(), cmap='CMRmap_r', vmax=high, extend='both')
    cbar = plt.colorbar(cs, shrink=0.5, ticks=np.arange(low, high, (high-low)/25), label = label)

    # Add coastlines and country borders
    ax.coastlines()
    ax.add_feature(cf.BORDERS)

    # Define the xticks for longitude
    ax.set_xticks(np.arange(-180,181,60), crs=ccrs.PlateCarree())
    lon_formatter = cticker.LongitudeFormatter()
    ax.xaxis.set_major_formatter(lon_formatter)

    # Define the yticks for latitude
    ax.set_yticks(np.arange(-80,81,20), crs=ccrs.PlateCarree())
    lat_formatter = cticker.LatitudeFormatter()
    ax.yaxis.set_major_formatter(lat_formatter)
    return fig, ax 
```

```python
global_figure(values_yr_percent_cell, 0, 3, 'Re-precipitation from source in %')
```

```python
# First we compute yearly evaporation and precipitation
E_per_year_month = np.load(link_link_path+"\\E_per_year_per_month_2001_2018.npy") # m3
P_per_year_month = np.load(link_link_path+"\\P_per_year_per_month_2001_2018.npy") # m3
E_year           = np.sum(E_per_year_month, axis=(0,1))/len(E_per_year_month[:,0,0,0]) # m3
P_year           = np.sum(P_per_year_month, axis=(0,1))/len(P_per_year_month[:,0,0,0]) # m3
A_gridcell       = np.load(link_link_path+"\\Further Data\\A_gridcell.npy") # m2
E_year_mm        = E_year / A_gridcell * 1000
P_year_mm        = P_year / A_gridcell * 1000

# evaporation from the source cell
global_figure(E_year_mm, 0, 3200, 'evaporation (mm/year)')

# precipitation in the source cell
global_figure(P_year_mm, 0, 5000, 'precipitation (mm/year)')
```

Using the equations in Van der Ent et al. (2010) and Van der Ent & Savenije (2011), regional metrics can be computed.

```python
# calculate regional recycling metrics

# distance over which recycling takes places within the grid cell
gridcell = 1.5 # degree
dg = 111089.56 # length of 1 degree latitude [m]
delta_x = gridcell * cos((closest_value_coordinates['lat']) * math.pi/180) * dg;

# evaporation from grid cell 
E = np.float64(E_year_mm[ind_lat,ind_lon])

# precipitation from grid cell
P = np.float64(P_year_mm[ind_lat,ind_lon])

# regional evaporation recycling ratio (fraction)
epsilon_r = np.float64(values_yr_percent_cell[ind_lat,ind_lon]) / 100

# regionally recycled water (mm/yr)
E_r = E * epsilon_r
P_r = E_r

# regional precipitation recycling ratio
rho_r =  P_r / P 

# length scale of evaporation recycling
lambda_epsilon = - delta_x  / (np.log(1-2*epsilon_r))

# length scale of precipitation recycling
lambda_rho = - delta_x  / (np.log(1-2*rho_r))

# print all results
print(f'E = \t\t{E:.2f} mm/year')
print(f'P = \t\t{P:.2f} mm/year')
print(f'epsilon_r = \t{epsilon_r*100:.2f} %')
print(f'E_r = \t\t{E_r:.2f} mm/year')
print(f'P_r = \t\t{P_r:.2f} mm/year')
print(f'rho_r = \t{rho_r*100:.2f} %')
print(f'delta_x = \t{delta_x/1000:.2f} km')
print(f'lambda_epsilon= {lambda_epsilon/1000:.2f} km')
print(f'lambda_rho = \t{lambda_rho/1000:.2f} km')
```

## 2. Using single grid cell moisture recycling metrics to estimate moisture recyling in a bigger region

The length scales of the moisture recycling can be used to estimate the recycling in a larger region, in this case the east coast of the US, assuming that the moisture recycling process is relatively constant. In order to do so we just need to estimate the length a water particle would travel on the east coast of the US

as moisture is mainly driven by wind, we consider the direction:
>![direction wind](Figures\direction_mositure_recycling.png)

and take the distance of the catchment (roughly) in this direction: towards the north east 
>![measured distance](Figures\distance_mositure_recycling.png)



```python
# exercise, regional recycling in east coast US based single grid cell info

# length a water particle travels on the east coast of the US
delta_x_US =  450_000 # km

# regional evaporation recycling ratio in Suriname
epsilon_r_US = (delta_x_US + lambda_epsilon * np.exp(- delta_x_US / lambda_epsilon) - lambda_epsilon)/ delta_x_US

# regional precipitation recycling ratio in Suriname
rho_r_US = (delta_x_US + lambda_rho * np.exp(- delta_x_US / lambda_rho) - lambda_rho)/ delta_x_US

print(f'delta_x_US =\t{delta_x_US/1000} km')
print(f'epsilon_r_US =\t{epsilon_r_US*100:.2f} %')
print(f'rho_r_US = \t{rho_r_US*100:.2f} %')
```
