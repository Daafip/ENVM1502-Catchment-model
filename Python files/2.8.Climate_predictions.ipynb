```python
import xarray as xr
import matplotlib.pyplot as plt
import cdsapi
import glob
import pandas as pd
import numpy as np
import os

import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from cartopy.feature.nightshade import Nightshade
```

```python
c = cdsapi.Client()

experiments = ['ssp5_3_4os','ssp5_8_5']
models     = ["ec_earth3",'ipsl_cm6a_lr']  
for index in range(len(experiments)):
    c.retrieve(
        'projections-cmip6',
            {
            'format': 'zip',
            'experiment': experiments[index],
            'variable': 'precipitation',
            'model': models[index],
            'year': [str(x) for x in list(np.linspace(2050,2099,50,dtype=int))], # all between 2050 & 2099
            'month': [str(x) for x in list(np.linspace(1,12,12,dtype=int))],     # all
            'day': [str(x) for x in list(np.linspace(1,31,31,dtype=int))],       # all    
            'area': [45, -77, 40,-70],                                           # bounding box of chosen region
            'temporal_resolution': 'daily'
            },
        f'{data_folder}\\Climate Predictions\\download{index}.zip')
```

```python
# This code shows one map
ifile = glob.glob(f'{data_folder}\\Climate Predictions\\*.nc')
print(ifile)
ds_534 = xr.open_mfdataset(ifile[0]) # open dataset as x-array
ds_585 = xr.open_mfdataset(ifile[1]) # open dataset as x-array
```

```python
ds_534
```

```python
lst_series_out = []
names = ["EC-Earth3_ssp534","IPSL-CM6A-LR_ssp585"]
fig, ax = plt.subplots(1)
for index, ds in enumerate([ds_534,ds_585]):
    mean_lat = ds.mean(dim="lat")
    mean_lat_lon = mean_lat.mean(dim="lon")
    series_out = mean_lat_lon.pr.to_pandas() * 86400
    series_out.plot(ax=ax, marker=".",lw=0,label=names[index])
    ax.set_title("daily precipitation values")
    ax.legend()
    lst_series_out.append(series_out)
```

```python
fig, ax = plt.subplots(1)
for index, series_out in enumerate(lst_series_out):    
    df = pd.DataFrame(data=series_out)
    df.resample("m").sum().plot(ax=ax,marker=".",label=names[index],zorder=1-index)
    ax.set_ylabel("mm/month")
    ax.set_title("monthly sum precipitation for two projections")
ax.legend(names)
```

```python
fig, ax = plt.subplots(1)
for index, series_out in enumerate(lst_series_out):    
    df = pd.DataFrame(data=series_out)
    df.resample("y").sum().plot(ax=ax,marker=".",label=names[index],zorder=1-index)
    ax.set_ylabel("mm/month")
    ax.set_title("yearly sum precipitation for two projections")
ax.legend(names)
```
